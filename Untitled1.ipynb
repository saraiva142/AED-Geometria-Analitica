```python
import matplotlib.pyplot as plt
import numpy as np
import re

def plot_elipse_from_equation(equation, ponto=None):
    match = re.match(r"(\d+)x\^2 \+ (\d+)y\^2 = (\d+)", equation)
    if not match:
        print("Formato da equação inválido.")
        return

    A, B, C = map(float, match.groups())

    if A <= 0 or B <= 0:
        print("A equação não representa uma elipse válida.")
        return

    h = 0
    k = 0

    a = np.sqrt(C / A)
    b = np.sqrt(C / B)

    if np.isnan(a) or np.isnan(b):
        print("A equação não representa uma elipse válida.")
        return

    c = np.sqrt(abs(a**2 - b**2))
    F1 = (h, k - c) if A > B else (h - c, k)
    F2 = (h, k + c) if A > B else (h + c, k)

    theta = np.linspace(0, 2 * np.pi, 100)
    x = h + a * np.cos(theta)
    y = k + b * np.sin(theta)

    plt.plot(x, y, label='Elipse')
    plt.scatter([F1[0], F2[0]], [F1[1], F2[1]], color='red', label='Focos')

    if ponto:
        pertence_elipse = (ponto[0]**2) / (a**2) + (ponto[1]**2) / (b**2) == 1
        plt.scatter(ponto[0], ponto[1], color='green' if pertence_elipse else 'black', label='Ponto')

        if pertence_elipse:
          print(f'O ponto {ponto} pertence à elipse')
        else:
          print(f'O ponto {ponto} não pertence à elipse')

    plt.title(f'Elipse a partir da Equação: {equation}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

    print(f'Coordenadas dos Focos:')
    print(f'Foco 1 (F1): {F1}')
    print(f'Foco 2 (F2): {F2}')
    print(f'Semi-eixo maior (a): {a}')
    print(f'Semi-eixo menor (b): {b}')

equation = "4x^2 + 1y^2 = 16"
ponto = (2, 0)
plot_elipse_from_equation(equation, ponto)
```

**Entrada Com Foco(s) e Ponto extremo**

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_elipse_from_foci_and_point(F1, F2, P, ponto=None):
    h = (F1[0] + F2[0]) / 2
    k = (F1[1] + F2[1]) / 2

    c = np.sqrt((F1[0] - F2[0])**2 + (F1[1] - F2[1])**2) / 2
    a = np.sqrt((P[0] - h)**2 + (P[1] - k)**2)

    b = np.sqrt(a**2 - c**2)

    theta = np.linspace(0, 2 * np.pi, 100)
    x = h + a * np.cos(theta)
    y = k + b * np.sin(theta)

    if ponto:
        pertence_elipse = (ponto[0]**2) / (a**2) + (ponto[1]**2) / (b**2) == 1
        plt.scatter(ponto[0], ponto[1], color='green' if pertence_elipse else 'black', label='Ponto')

        if pertence_elipse:
          print(f'O ponto {ponto} pertence à elipse')
        else:
          print(f'O ponto {ponto} não pertence à elipse')


    plt.plot(x, y, label='Elipse')
    plt.scatter([F1[0], F2[0]], [F1[1], F2[1]], color='red', label='Focos')
    plt.scatter(P[0], P[1], color='green', label='Ponto')

    plt.title('Elipse a partir dos Focos e Ponto Extremo')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

    print(f'Coordenadas dos Focos:')
    print(f'Foco 1 (F1): {F1}')
    print(f'Foco 2 (F2): {F2}')
    print(f'Semi-eixo maior (a): {a}')
    print(f'Semi-eixo menor (b): {b}')
    print(f'Centro (h, k): ({h}, {k})')

F1 = (0, -2)
F2 = (0, 2)
PontoExtremo = (3, 0)
ponto = (-3, 0)

plot_elipse_from_foci_and_point(F1, F2, PontoExtremo, ponto)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import re

def ponto_pertence_parabola(A, ponto):
    x, y = ponto

    return y == -A * x**2

def plot_parabola_from_equation(equation, ponto):
    match = re.match(r"(\d+)x\^2 \+ (\d+)y = 0", equation)
    if not match:
        print("Formato da equação inválido.")
        return

    A, B = map(float, match.groups())

    if A == 0 or B == 0:
        print("A equação não representa uma parábola válida.")
        return

    h = 0
    k = 0

    pertence = ponto_pertence_parabola(A, ponto)

    x = np.linspace(-10, 10, 1000)
    y = -A * x**2 / B

    plt.plot(x, y, label='Parábola')

    plt.scatter(*ponto, color='red', label='Ponto P(x, y)')

    plt.scatter(h, k, color='green', label='Vértice')

    plt.title(f'Parábola a partir da Equação: {equation}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f'Vértice da Parábola:')
    print(f'Vertice: ({h}, {k})')

    if pertence:
        print(f'O ponto P({ponto[0]}, {ponto[1]}) pertence à parábola.')
    else:
        print(f'O ponto P({ponto[0]}, {ponto[1]}) não pertence à parábola.')

equation = "1x^2 + 4y = 0"
ponto = (0, 0)
plot_parabola_from_equation(equation, ponto)
```

* Entrada apenas com focos e vértice

```python
import numpy as np
import matplotlib.pyplot as plt

def ponto_pertence_parabola(a, h, k, x, y):
    return y == a * (x - h)**2 + k

def criar_parabola(vertice, foco):
    h, k = vertice
    f, g = foco

    if h == f:
        a = 1 if g > k else -1
        direcao = "para cima" if g > k else "para baixo"
        equacao = f"y = {a}(x - {h})^2 + {k}"
        x = np.linspace(h - 5, h + 5, 1000)
        y = a * (x - h)**2 + k
    elif k == g:
        a = 1 if f > h else -1
        direcao = "para a direita" if f > h else "para a esquerda"
        equacao = f"y = {a}(x - {h})^2 + {k}"
        y = np.linspace(k - 5, k + 5, 1000)
        x = a * (y - k)**2 + h
    else:
        raise ValueError("Vértice e foco não definem uma parábola válida.")

    print(f"A parábola {equacao} abre {direcao}.")

    ponto_x = 6
    ponto_y = 5
    pertence = ponto_pertence_parabola(a, h, k, ponto_x, ponto_y)
    if pertence:
        print(f'O ponto P({ponto_x}, {ponto_y}) pertence à parábola.')
    else:
        print(f'O ponto P({ponto_x}, {ponto_y}) não pertence à parábola.')

    plt.plot(x, y, label=equacao)
    plt.scatter([h, f], [k, g], color='red')
    plt.scatter(ponto_x, ponto_y, color='green')
    plt.title('Parábola')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.show()

vertice = (5, 4)
foco = (8, 4)
criar_parabola(vertice, foco)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def criar_h(vertice1, vertice2, foco, ponto):
    x1, y1 = vertice1
    x2, y2 = vertice2
    fx, fy = foco

    h = (x1 + x2) / 2
    k = (y1 + y2) / 2

    a = np.sqrt((x1 - h)**2 + (y1 - k)**2)
    b = np.sqrt((x2 - h)**2 + (y2 - k)**2)
    c = np.sqrt((fx - h)**2 + (fy - k)**2)

    e = c / a

    if x1 == x2:
        x = np.linspace(h - a - 1, h + a + 1, 400)
        y_pos = k + b * np.sqrt(1 + ((x - h)**2 / a**2))
        y_neg = k - b * np.sqrt(1 + ((x - h)**2 / a**2))
        plt.plot(x, y_pos, label=f"Hipérbole: e={e}")
        plt.plot(x, y_neg, color='blue')
    elif y1 == y2:
        y = np.linspace(k - b - 1, k + b + 1, 400)
        x_pos = h + a * np.sqrt(1 + ((y - k)**2 / b**2))
        x_neg = h - a * np.sqrt(1 + ((y - k)**2 / b**2))
        plt.plot(x_pos, y, label=f"Hipérbole: e={e}")
        plt.plot(x_neg, y, color='blue')
    else:
        raise ValueError("Vértice e foco não definem uma hipérbole válida.")

    print(f"A hipérbole é definida pelos pontos dados.")
    print(f"Centro da hipérbole: ({h}, {k})")
    print(f"Outro foco da hipérbole: ({2 * h - fx}, {2 * k - fy})")

    plt.scatter([x1, x2, fx, h, 2 * h - fx], [y1, y2, fy, k, 2 * k - fy], color='red')

    ponto_x, ponto_y = ponto
    if (ponto_x - h)**2 / a**2 - (ponto_y - k)**2 / b**2 == 1:
        plt.scatter(ponto_x, ponto_y, color='green', label='Pertence')
    else:
        plt.scatter(ponto_x, ponto_y, color='orange', label='Não Pertence')

    plt.title('Hipérbole e Ponto P(x, y)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.show()

A1 = (1, -2)
A2 = (5, -2)
foco = (6, -2)
ponto = (2, -2)

criar_h(A1, A2, foco, ponto)
```
